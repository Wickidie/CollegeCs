```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img_list = [
    "city_smoothing.jpg",
    "train_smoothing.jpg",
    "butterfly_sharpening.jpg",
    "zaha_sharpening.jpg"
]
plt.figure(figsize=(16,9))

def praktek_1(img_list):
    # Median
    img_ori = cv.imread(img_list[0])
    img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
    img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
    plt.title("Original")
    plt.imshow(img_ori)
    plt.figure(figsize=(16,9))
    plt.title("Grayscale")
    plt.imshow(img_gray, cmap="gray")
    plt.figure(figsize=(16,9))

    img_smooth = cv.blur(img_gray, (5, 5))
    plt.title("Average Kernel Size 5")
    plt.imshow(img_smooth, cmap="gray")
    plt.figure(figsize=(16,9))
    img_smooth = cv.medianBlur(img_gray, 5)
    plt.title("Median Kernel Size 5")
    plt.imshow(img_smooth, cmap="gray")
    plt.figure(figsize=(16,9))

    # Median then average
    img_ori = cv.imread(img_list[1])
    img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
    img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
    plt.title("Original")
    plt.imshow(img_ori)
    plt.figure(figsize=(16,9))
    plt.title("Grayscale")
    plt.imshow(img_gray, cmap="gray")
    plt.figure(figsize=(16,9))
    

    img_smooth = cv.blur(img_gray, (7, 7))
    plt.title("Average Kernel Size 7")
    plt.imshow(img_smooth, cmap="gray")
    plt.figure(figsize=(16,9))
    img_smooth = cv.medianBlur(img_gray, 9)
    plt.title("Median Kernel Size 9")
    plt.imshow(img_smooth, cmap="gray")
    plt.figure(figsize=(16,9))

def praktek_2(img_list):
    kernel_top = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    kernel_right = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    kernel_bottom = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
    kernel_left = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])

    for i in range(2, 4):
        # Laplacian
        img_ori = cv.imread(img_list[i])
        img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
        img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)

        img_laplacian = cv.Laplacian(img_gray, ddepth=-1)
        plt.title("Original")
        plt.imshow(img_ori)
        plt.figure(figsize=(16,9))
        plt.title("Gray")
        plt.imshow(img_gray, cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Laplacian")
        plt.imshow(img_laplacian, cmap="gray")
        plt.figure(figsize=(16,9))

        # Prewitt
        img_ori = cv.imread(img_list[i])
        img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
        img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)

        img_prewitt_top = cv.filter2D(img_gray, -1, kernel_top)
        img_prewitt_right = cv.filter2D(img_gray, -1, kernel_right)
        img_prewitt_bottom = cv.filter2D(img_gray, -1, kernel_bottom)
        img_prewitt_left = cv.filter2D(img_gray, -1, kernel_left)
        img_prewitt_2way = img_prewitt_top + img_prewitt_right
        img_prewitt_4way = img_prewitt_top + img_prewitt_right + img_prewitt_bottom + img_prewitt_left

        plt.title("Prewitt Top")
        plt.imshow(img_prewitt_top,cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Prewitt Right")
        plt.imshow(img_prewitt_right,cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Prewitt Bottom")
        plt.imshow(img_prewitt_bottom,cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Prewitt Left")
        plt.imshow(img_prewitt_left,cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Prewitt 2way")
        plt.imshow(img_prewitt_2way,cmap="gray")
        plt.figure(figsize=(16,9))
        plt.title("Prewitt 4way")
        plt.imshow(img_prewitt_4way,cmap="gray")
        plt.figure(figsize=(16,9))

praktek_1(img_list)
praktek_2(img_list)
```
