```python
import os
import io
import PIL
import numpy as np
import tensorflow as tf
import PySimpleGUI as sg
import matplotlib.pyplot as plt

from PIL import Image
from imutils import paths
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

from roboflow import Roboflow
rf = Roboflow(api_key="2p3nu1nFzRHYX26AJYXx")
project = rf.workspace().project("dog-breeds-classifications")
model = project.version(1).model

def predictDog(dog_path):

    # infer on a local image
    res = model.predict(dog_path).json()
    first_prediction_dict = res['predictions'][0]

    # Accessing specific elements within the first prediction dictionary
    class_name = first_prediction_dict['top']
    class_id = first_prediction_dict['predictions'][0]['class_id']
    confidence = first_prediction_dict['confidence'] * 100

    # Displaying the extracted information
    return class_name, confidence

# dog_path = "n02106166_1032.jpg"
# predictDog(dog_path)

font = ("Roboto", 20)
sg.theme('Dark Blue')

cwd = os.path.expanduser('~/Downloads')
file_types = [("All files (*.*)", "*.*")]

allign_center = [  
    # [
    #     sg.Text("Image Sample"),
    # ],
    # [
    #     sg.Image(key="-IMAGE_EXAMPLE_1-"),
    #     sg.Image(key="-IMAGE_EXAMPLE_3-"),
    #     sg.Image(key="-IMAGE_EXAMPLE_2-")
    # ],
    [
        sg.Text("Dog Breed : ", key="-DOG_BREED-"),
    ],
    [
        sg.Text("Confident : ", key="-CONFIDENT-"),
    ],
    [
        sg.Image(key="-IMAGE_PREDICTION-")
    ],
    [
        sg.Text("Image File Path"),
        sg.Input(size=(25, 1), enable_events=True, key="-FILE_PATH-"),
        sg.FileBrowse(initial_folder=cwd, file_types=file_types),
    ],
    [
        sg.Button("Submit Image", font=("Roboto", 14))
    ],
]

main_menu_window = [
    [sg.Push(), sg.Column(allign_center, element_justification='c')]
]

window = sg.Window("Dog Breed Classification", main_menu_window)
while True:
    event, values = window.read()
    if event in (sg.WIN_CLOSED, "Exit"):
        break
    elif event == 'Submit Image':
        filename = values["-FILE_PATH-"]
        print(filename)
        dog_breed, confident = predictDog(filename)
        dog_breed = "Dog Breed : " + dog_breed
        confident = "Confident : " + str(round(confident, 2)) + "%"
        if os.path.exists(filename):
            image = Image.open(values["-FILE_PATH-"])
            image.thumbnail((600, 600))
            bio = io.BytesIO()
            image.save(bio, format="PNG")
            window["-IMAGE_PREDICTION-"].update(data=bio.getvalue())
            window["-DOG_BREED-"].update(dog_breed)
            window["-CONFIDENT-"].update(confident)
            # window["-IMAGE_EXAMPLE_1-"].update(data=bio.getvalue())
            # window["-IMAGE_EXAMPLE_2-"].update(data=bio.getvalue())
            # window["-IMAGE_EXAMPLE_3-"].update(data=bio.getvalue())

window.close()
```
