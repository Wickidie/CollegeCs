```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

img_list = [
    "Starry_Night_Ori.jpg"
]
# plt.figure(figsize=(16, 9))

# Tresholding
# Smongko
img_ori = cv.imread(img_list[0])
img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
plt.title("Semangka Original")
plt.imshow(img_ori)
plt.figure()

img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
plt.title("Semangka Grayscale")
plt.imshow(img_gray, cmap="gray")
plt.figure()
# plt.title("Histogram Intensitas Semangka Grayscale")
# plt.hist(img_gray.ravel(), 255, [0, 256])
# plt.figure()

ret, thresh = cv.threshold(img_gray, 126, 255, cv.THRESH_BINARY)
plt.title("Binary")
plt.imshow(thresh, cmap="gray")
plt.figure()

thresh = cv.adaptiveThreshold(img_gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 7, 2)
plt.title("Adaptive Mean")
plt.imshow(thresh, cmap="gray")
plt.figure()

thresh = cv.adaptiveThreshold(img_gray, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 7, 2)
plt.title("Adaptive Gaussian")
plt.imshow(thresh, cmap="gray")
plt.figure()

ret, thresh = cv.threshold(img_gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
plt.title("Otsu")
plt.imshow(thresh, cmap="gray")
plt.figure()

plt.show()
```
