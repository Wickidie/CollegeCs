```python
import cv2 as cv
import matplotlib.pyplot as plt

img_list = [
    "Histeq_Gambar I.jpg",
    "Histeq_Gambar II.jpg",
    "Histeq_Gambar III.jpg",
    "Treshold_Gambar I.jpg",
    "Treshold_Gambar II.jpg"
]

# Tresholding
# Smongko
img_ori = cv.imread(img_list[3])
img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
plt.title("Semangka Original")
plt.imshow(img_ori)
plt.figure()

img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
plt.title("Semangka Grayscale")
plt.imshow(img_gray, cmap="gray")
plt.figure()
plt.title("Histogram Intensitas Semangka Grayscale")
plt.hist(img_gray.ravel(), 255, [0, 256])
plt.figure()

ret, thresh = cv.threshold(img_gray, 250, 256, cv.THRESH_BINARY)
plt.title("Semangka Greyscale Threshold m=250")
plt.imshow(thresh, cmap="gray")
plt.figure()

# Dori
img_ori = cv.imread(img_list[4])
img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
plt.title("Dori Original")
plt.imshow(img_ori)
plt.figure()

img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
plt.title("Dori Grayscale")
plt.imshow(img_gray, cmap="gray")
plt.figure()
r, g, b = cv.split(img_ori)
r = np.zeros(r.shape, np.uint8)
img_gb = cv.merge([r, g, b])

img_gray = cv.cvtColor(img_gb, cv.COLOR_RGB2GRAY)
plt.title("Dori Grayscale warna Red dihilangkan")
plt.imshow(img_gray, cmap="gray")
plt.figure()
plt.title("Histogram Intensitas Dori Grayscale")
plt.hist(img_gray.ravel(), 255, [0, 256])
plt.figure()

ret, thresh = cv.threshold(img_gray, 49, 256, cv.THRESH_BINARY)
plt.title("Dori Grayscale warna Red dihilangkan Threshold m=49")
plt.imshow(thresh, cmap="gray")
plt.figure()

plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img_list = [
    "Histeq_Gambar I.jpg",
    "Histeq_Gambar II.jpg",
    "Histeq_Gambar III.jpg",
    "Treshold_Gambar I.jpg",
    "Treshold_Gambar II.jpg"
]

# Histogram Equalization
for i in range(0, 3):
    img_ori = cv.imread(img_list[i])
    img_ori = cv.cvtColor(img_ori, cv.COLOR_BGR2RGB)
    plt.imshow(img_ori)
    plt.figure()
    plt.hist(img_ori.ravel(), 255, [0, 256])
    plt.figure()

    img_gray = cv.cvtColor(img_ori, cv.COLOR_RGB2GRAY)
    img_smooth = cv.equalizeHist(img_gray)
    plt.imshow(img_smooth, cmap="gray")
    plt.figure()
    plt.hist(img_smooth.ravel(), 255, [0, 256])
    plt.figure()    


```
